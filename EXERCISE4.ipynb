```python
hourly_wage = float(input("Enter hourly wage: "))
hours = float(input("Enter total working hours: "))
if hours > 40:
    wage = (40 * hourly_wage) + ((hours - 40) * hourly_wage * 1.5)
else:
    wage = hours * hourly_wage
print(f"Total wage: {wage:.2f}")  
```

```python
a, b, c = map(int, input("Enter three sides: ").split())
sides = sorted([a, b, c])
if sides[0]**2 + sides[1]**2 == sides[2]**2:
    print(f"{a}, {b}, {c} are the sides of a right triangle")
else:
    print(f"{a}, {b}, {c} are not the sides of a right triangle")  
```

```python
students = {
    "Frank": 65,
    "Annie": 52,
    "Ben": 71,
    "Tommy": 80,
    "Jenson": 99,
    "Lisa": 25
}
for name, score in students.items():
    if score >= 80:
        grade = 'A'
    elif score >= 60:
        grade = 'B'
    elif score >= 50:
        grade = 'C'
    elif score >= 40:
        grade = 'D'
    else:
        grade = 'F'
    print(f"{name}: Score = {score}, Grade = {grade}")  
```

```python
n = int(input("Enter number of Fibonacci terms n: "))
if n <= 0:
    fib = []
elif n == 1:
    fib = [0]
else:
    fib = [0, 1]
    while len(fib) < n:
        fib.append(fib[-1] + fib[-2])
print(":", fib) 
```
