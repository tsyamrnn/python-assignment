#### (c) What is the new date after adding 2 years, 5 months and 15 days to 17th November 2020?

#### (h)  Supposed that the 2024 Summer Olympics opening ceremony is at 8 PM Paris local time. Find the local time for the spectators in following cities to watch the event live.
Kuala Lumpur, Seoul, Muscat, Ontario, Calcutta

```python
import arrow
paris_time = arrow.get("2024-07-26 20:00", "YYYY-MM-DD HH:mm").replace(tzinfo="Europe/Paris")

cities = {
    "Kuala Lumpur": "Asia/Kuala_Lumpur",
    "Seoul": "Asia/Seoul",
    "Muscat": "Asia/Muscat",
    "Ontario": "America/Toronto",
    "Calcutta": "Asia/Kolkata"
}

for city, zone in cities.items():
    print(city, ":", paris_time.to(zone).format("YYYY-MM-DD HH:mm"))
```

#### Question 3

#### (a) David works on weekdays and gets paid $120 a day. Calculate David’s salary for a given month and year

```python
import calendar

year = 2024
month = 8   
pay_per_day = 120
salary = 0

for day in range(1, calendar.monthrange(year, month)[1] + 1):
    weekday = calendar.weekday(year, month, day)
    if weekday < 5:   
        salary += pay_per_day

print("Salary for", month, "/", year, "is $", salary)
```

#### (b) David works on weekdays and gets paid $120 a day but the pay doubles if the week day is a holiday. Calculate David’s salary for a given month, year and a list of holidays

```python
import calendar

year = 2024
month = 8
pay_per_day = 120
salary = 0

holidays = [8, 31]   
for day in range(1, calendar.monthrange(year, month)[1] + 1):
    weekday = calendar.weekday(year, month, day)
    if weekday < 5:  
        if day in holidays:
            salary += pay_per_day * 2  
        else:
            salary += pay_per_day

print("Salary for", month, "/", year, "with holidays is $", salary)
```

#### (g) Copy the following data into a csv file experties.csv. Then read the data and convert all texts into title case and write it into another csv file experties_title_case.csv

```python
import csv

data = [
    ["mark thompson", "associate attorney", "johnson & yun associates"],
    ["alan johnson", "business executive", "parkland holdings"],
    ["jenny clerk", "financial advisor", "public bank bhd."],
    ["timothy white", "investment analyst", "fieldman financial group"]
]

with open("experties.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["name","occupation","association"])
    writer.writerows(data)

# convert to title case
with open("experties.csv", "r") as f:
    reader = csv.reader(f)
    rows = list(reader)

with open("experties_title_case.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for row in rows:
        new_row = [x.title() for x in row]
        writer.writerow(new_row)
```

#### (c)  Calculate the grade of each student (score ≥ 80 for A, 60 ≤ score < 80 for B, 40 ≤ score < 60 for C, and score < 40 for E). Write all data in students.csv to student_grade.csv and add the grades as the fourth column of student_grade.csv.

```python
import csv

def grade(score):
    if score >= 80:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 40:
        return "C"
    else:
        return "E"

with open("student_grade.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name","ID","Score","Grade"])
    writer.writerow(["Ali","S001",85,grade(85)])
    writer.writerow(["Fatimah","S002",72,grade(72)])
    writer.writerow(["John","S003",60,grade(60)])
    writer.writerow(["Aisyah","S004",45,grade(45)])
    writer.writerow(["Tom","S005",30,grade(30)])
    writer.writerow(["Sara","S006",95,grade(95)])
    writer.writerow(["Adam","S007",70,grade(70)])
    writer.writerow(["Maya","S008",55,grade(55)])
    writer.writerow(["Lee","S009",40,grade(40)])
    writer.writerow(["Zara","S010",88,grade(88)])
```
