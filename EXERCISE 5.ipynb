### 3. Use | a) aloop b) the map function
on the function written in Q2(d) to convert the following temperatures 125.70F, 275.40F, 300.00F, 37.480F, 0.00F, −180.450F to Celsius and print the results in tabular form.

```python
def celsius(f):
    return (f - 32) * 5/9


temps_f = [32, 68, 86, 104, 212]


results = list(map(celsius, temps_f))

print("Results using map():")
print("Fahrenheit\tCelsius")
print("-" * 25)

for f, c in zip(temps_f, results):
    print(f"{f}\t\t{c:.2f}")
```

### 8. Explain why each of the following calls to f is illegal CHAPTER 6.2.1

```python
def f(a, b, c = 0, d = 0):
    return a, b, c, d
```

(a) **f(a = 1)** :Illegal because `b` is a required argument and it is missing.

(b) **f(1, 2, 3, 4, 5)**: Illegal because the function only takes 4 arguments, but this call gives 5.

(c) **f(b = 1, 2)**: Illegal because you can't put a plain number after a named one (like `b=1`). Also, `a` is missing.

(d) **f(b = 1, c = 2)**: Illegal because `a` is required and it was never given a value.

(e) **f(1, b = 1, 2)**: Illegal because after using a keyword argument (`b=1`), all arguments that follow must also be named.

### 9.
##### Create the list of integers with the following pattern
[1, 4, 2, 4, 2, 4, ... , 2, 4, 1]
by returning it from a function. Note that the list starts and ends with 1, has odd-index elements
equal to 4 and even-indexed elements equal to 2. The function receives the length of the list as the
input. The length must by an odd number, if it is even, return an empty list.

```python
def create_pattern_list(length):
    if length % 2 == 0:
        return []
    my_list = [1]
    for i in range(1, length - 1):
        if i % 2 == 1:
            my_list.append(4) 
        else:
            my_list.append(2) 
    my_list.append(1)
    return my_list
```

```python
result = create_pattern_list(7)
print(result)
```

```python
print(create_pattern_list(4))
```

```python
print(create_pattern_list(11))
```
